## MyList ë§Œë“¤ê¸°

> ì´ì œ ë°°ìš´ ê²ƒì„ í™œìš©í•˜ì—¬ ë‚˜ë§Œì˜ ë¦¬ìŠ¤íŠ¸ ê°ì²´ë¥¼ ë§Œë“¤ ìˆ˜ ìˆìŠµë‹ˆë‹¤. 
>
> `class MyList:`
>

```
* í´ë˜ìŠ¤ ë³€ìˆ˜
data : ë¹„ì–´ ìˆëŠ” ë¦¬ìŠ¤íŠ¸

* ë©”ì„œë“œ 
append() : ê°’ì„ ë°›ì•„ data ì— ì¶”ê°€í•©ë‹ˆë‹¤. ë¦¬í„´ ê°’ì€ ì—†ìŠµë‹ˆë‹¤.
pop() : ë§ˆì§€ë§‰ì— ìˆëŠ” ê°’ì„ ì‚­ì œí•˜ê³ , í•´ë‹¹ ê°’ì„ ë¦¬í„´í•©ë‹ˆë‹¤.
reverse() : ì œìë¦¬ì—ì„œ ë’¤ì§‘ê³  ë¦¬í„´ ê°’ì€ ì—†ìŠµë‹ˆë‹¤.
count(x) : xì˜ ê°œìˆ˜ë¥¼ ë¦¬í„´í•©ë‹ˆë‹¤.

__repr__() : ex) 'ë‚´ ë¦¬ìŠ¤íŠ¸ì—ëŠ” [1, 2, 3] ì´ ë‹´ê²¨ìˆë‹¤.'
```

````python
# ì•„ë˜ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class MyList:
    data = []
    
    def append(self, element):
        self.data += [element]
    
    def pop(self):
        p = self.data[-1]
        self.data = self.data[:-1]
        return p
    
    def reverse(self):
        self.data = self.data[::-1]
        
    def count(self):
        return len(self.data)
    
    def clear(self):
        self.data = []
        
    def __repr__(self):
        return f'ë‚´ ë¦¬ìŠ¤íŠ¸ì—ëŠ” {self.data} ê°€ ë‹´ê²¨ìˆë‹¤.'
````

## ì‹¤ìŠµ (ì¢…í•©)

> ì‚¬ì‹¤ ì´ì „ì— ì‘ì„±í•œ MylistëŠ” ì™„ë²½í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. 
>
> í•œë²ˆ ì œëŒ€ë¡œ ëœ ìë£Œêµ¬ì¡°ë¥¼ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤. 
>
> `Stack` í´ë˜ìŠ¤ë¥¼ ê°„ëµí•˜ê²Œ êµ¬í˜„í•´ë´…ì‹œë‹¤.

> [Stack](https://ko.wikipedia.org/wiki/%EC%8A%A4%ED%83%9D) : ìŠ¤íƒì€ LIFO(Last in First Out)ìœ¼ë¡œ êµ¬ì¡°í™”ëœ ìë£Œêµ¬ì¡°ë¥¼ ëœ»í•©ë‹ˆë‹¤.

1. `empty()`: ìŠ¤íƒì´ ë¹„ì—ˆë‹¤ë©´ Trueì„ ì£¼ê³ , ê·¸ë ‡ì§€ ì•Šë‹¤ë©´ Falseê°€ ëœë‹¤.

2. `top()`: ìŠ¤íƒì˜ ê°€ì¥ ë§ˆì§€ë§‰ ë°ì´í„°ë¥¼ ë„˜ê²¨ì¤€ë‹¤. ìŠ¤íƒì´ ë¹„ì—ˆë‹¤ë©´ Noneì„ ë¦¬í„´í•œë‹¤.

3. `pop()`: ìŠ¤íƒì˜ ê°€ì¥ ë§ˆì§€ë§‰ ë°ì´í„°ì˜ ê°’ì„ ë„˜ê²¨ì£¼ê³ , í•´ë‹¹ ë°ì´í„°ë¥¼ ì‚­ì œí•œë‹¤. ìŠ¤íƒì´ ë¹„ì—ˆë‹¤ë©´ Noneì„ ë¦¬í„´í•œë‹¤.

4. `push()`: ìŠ¤íƒì˜ ê°€ì¥ ë§ˆì§€ë§‰ ë°ì´í„° ë’¤ì— ê°’ì„ ì¶”ê°€í•œë‹¤. ë¦¬í„´ ê°’ì€ ì—†ë‹¤.

**ë‹¤ ì‘ì„±í–ˆë‹¤ë©´ __repr__ ì„ í†µí•´ ì˜ˆì˜ê²Œ ì¶œë ¥ê¹Œì§€ í•´ë´…ì‹œë‹¤.**

````python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ì‘ì„±í•´ì£¼ì„¸ìš”.
class Stack:
    name = 'ìë£Œêµ¬ì¡° - stack'
    def __init__(self):
        self.data = []  # ì´ê²Œ ëª…ì„¸ì˜ ìŠ¤íƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤.
        
    def push(self, *element):
        self.data += list(element)  # [1, 2, 3]
        return None
    
    def pop(self):
        if len(self.data) == 0:
            return None
        else:
            p = self.data[-1]
            self.data = self.data[:-1]
            return p
        
    def is_empty(self):
        if len(self.data) == 0:
            return True
        else:
            return False
    
    def top(self):
        if len(self.data) == 0:
            return None
        else:
            return self.data[-1]
        
    def __repr__(self):
        show = []
        show.append(' ___ ')
        for e in self.data:
            show.append(f'| {e} |\n')
        show.reverse()
        return ''.join(show)
````

````python
# ì—¬ê¸°ì— í´ë˜ìŠ¤ë¥¼ ì •ì˜í•˜ì„¸ìš”
class Circle:
    pi = 3.14
    def __init__(self, radius, coord_x=0, coord_y=0):
        self.r = radius
        self.x = coord_x
        self.y = coord_y
        
    def area(self):
        return Circle.pi * (self.r**2)
    
    def circumference(self):
        return 2*Circle.pi*self.r

    
    def center(self):
        return self.x, self.y
    
    def move(self, x, y):
        self.x, self.y = x, y
        return (self.x, self.y)


c1 = Circle(10, -1, 3)
print(c1.r, c1.x, c1.y)
c1.center()


````
